```python
def greetings(name):
    greeting = f'Hello GoodMorning {name}\nWelcome youre next task is parameters and arguments'
    return greeting
print(greetings("this is an argument"))
print(greetings("rushil"))
print(greetings(""))
```

```python
def area(r):
    pi = 3.14
    area = pi * r * r 
    return area 
print("area of the circle is :", area(5))
```

```python
def sum_of_numbers(n): # n is a parameter which can be any number, here we are passing 10 and 100 as arguments
    start = 0 
    for i in range (n+1): #n is 10 then range will be till 9, so +1 is added to include 10 in the range
        start += i        #here as i is the iteration it will take from 0 to 10 and add it to start variable which is initialized to 0
    return start
print("sum till following number is:", sum_of_numbers(10))
print("sum till following number is:", sum_of_numbers(100)) #argumments
```

```python
def odd_numbers(n):
    odd_nums = []
    for i in range(n+1):
        if i % 2!=0:
            odd_nums.append(i)
            
    return odd_nums
print("odd numbers till following number is:", odd_numbers(100))
        
```

```python
def calculate_area(Length, Breadth):
    area = Length * Breadth
    return area 
print("area of rectangle is:", calculate_area(10,20))
      
```

*args - allows to give multiple arguments for the parameter
**kwargs - allows to give multiple key - value arguments for parameter

```python
def your_order(total,*food_items, **bill ):
    print(f'You owe {total} for the following items:')
    for item in food_items:
        print(f'-{item}')
    print("Your bill details:")
    for key , value in bill.items():
        print(f' - {key}: {value}')

print(your_order(100,"pizza","burger","fries,coke", pizza=50, burger=30, fries=20, coke=10))
```
